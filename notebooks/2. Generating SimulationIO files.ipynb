# SimulationIO Shell Utilities

This notebook teaches you how to use bash utilities to create and look at your very own SimulationIO files. This is a bash notebook, so evaluating a cell with `shift+enter` runs one or more shell commands.

## Preliminaries:

Jupyterhub removes /usr/local/bin from the default path. We want that path so we add it back.

```bash
export PATH=/usr/local/bin:${PATH}
```

## Converting Output

The command 

`sio-convert-carpet-output` 

reads in one or more hdf5 files containing 3d Carpet output and outputs a new file which contains the same output and also the metadata generated by the SimulationIO library. The basic syntax is:

`sio-convert-carpet-output new-filename.s5 /path/to/target/files/*.h5`

Evaluate the following cell to convert the output for a cell-centred static neutron star (warning the output will be long):

```bash
sio-convert-carpet-output ./static-tov-cell.s5 ./data/static_tov_cell/output-0000/static_tov_cell_centred/*.h5
```

The file should now be present in your home directory:

```bash
ls -lh static-tov-cell.s5
```

## Printing Simulationio Metadata

Your conversion command should have output a ton of metadata. You can regenerate this metadata with the `sio-list` command:

```bash
sio-list ./static-tov-cell.s5
```
